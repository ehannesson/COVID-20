```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
```

```python
url = 'http://blog.lazd.net/coronadatascraper/'
```

```python
a = requests.get(url + './timeseries.csv')
decoded = a.content.decode('utf-8')

cr = csv.reader(decoded.splitlines(),delimiter=',')
my_list = list(cr)
```

```python
data = pd.DataFrame(my_list[1:], columns=my_list[0])
data = data[data.city == '']
data.drop('city', inplace=True, axis=1)
```

```python
byCounty = data[data.county != '']
byState = data[(data.state != '') & (data.county == '')]
byCountry = data[(data.state == '') & (data.county == '')]
countries = set(byCountry.country)
c, s = set(byCounty.country), set(byState.country)
print([(_, _ in countries) for _ in c])
print([(_, _ in countries) for _ in s])
```
