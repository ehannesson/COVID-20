```python
css_confirmed = pd.read_csv('csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')
css_confirmed.rename({'Province/State': 'state', 'Country/Region': 'country'}, inplace=True, axis=1)

who_data = pd.read_csv('who_covid_19_situation_reports/who_covid_19_sit_rep_time_series/who_covid_19_sit_rep_time_series.csv')
who_data
```

```python
by_country = css_confirmed.groupby('country').sum().drop(['Lat', 'Long'], axis=1)
```

```python
over_1000 = by_country[by_country['3/12/20']>1000]
over_1000_no_china = over_1000[over_1000.index != 'China']

plt.rcParams['figure.figsize'] = 23, 11
over_1000_no_china.T.plot()

plt.show()
```

```python
US = css_confirmed[css_confirmed.country == 'US'].groupby('state').sum().drop(['Lat', 'Long'], axis=1)
ax = US.loc['Utah'].plot()

ax.set_title('U')
plt.show()
```
